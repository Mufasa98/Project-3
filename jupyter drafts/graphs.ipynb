```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
from jupyterthemes import jtplot
#sets plot styles to match jupyter theme
jtplot.style('grade3')
```

```python
# import csvs
academics = pd.read_csv('../csvs/academics_clean.csv')
fafsa = pd.read_csv('../csvs/clean_fafsa.csv')
demographics= pd.read_csv('../csvs/demographics_clean.csv')
dependency = pd.read_csv('../csvs/dependency_df.csv')
gender_completion = pd.read_csv('../csvs/gender_completion_clean.csv')
income = pd.read_csv('../csvs/income_clean.csv')


```

```python
# Calculate the correlation coefficient and linear regression model 
#Get the correlation value 

correlation = round(st.pearsonr(academics['Completion Rate 4 year (American Indian)'],academics['SAT Average Overall'])[0],2)
print(f"The correlation between SAT scores and % AIAN Graduated Students is: {correlation}")

#linear regression values 
linear_reg = st.linregress(academics['Completion Rate 4 year (American Indian)'], academics['SAT Average Overall'])
line_eq = "y = " + str(round(linear_reg[0],2)) + "x + " + str(round(linear_reg[1],2))
#use y= mx +b for linear regression line.
y_mxb = academics['Completion Rate 4 year (American Indian)']*linear_reg[0]+linear_reg[1]

#Create scatter plot
plt.scatter(academics['Completion Rate 4 year (American Indian)'], academics['SAT Average Overall'])

#Plot the linear regression line on the scatter 
plt.plot(academics['Completion Rate 4 year (American Indian)'],y_mxb,color="red")

plt.xlabel('Completion Rate 4 year (American Indian)')
plt.ylabel('SAT Average Overall')
plt.annotate(line_eq,(20,800),fontsize=15,color="red")
plt.show()
```
